```python
#Making attribue names consistent with upper camel case
table.rename(columns={
    'gender': 'gender',
    'NationalITy': 'nationality',
    'PlaceofBirth': 'placeOfBirth',
    'StageID': 'stageID',
    'GradeID': 'gradeID',
    'SectionID': 'sectionID',
    'Topic': 'topic',
    'Semester': 'semester',
    'Relation': 'relation',
    'raisedhands': 'raisedHands',
    'VisITedResources': 'visitedResources',
    'AnnouncementsView': 'announcementsView',
    'Discussion': 'discussion',
    'ParentAnsweringSurvey': 'parentAnsweringSurvey',
    'ParentschoolSatisfaction': 'parentSchoolSatisfaction',
    'StudentAbsenceDays': 'studentAbsenceDays',
    'Class': 'class'
}, inplace=True)
```

```python
# Changing class to marks to avoid confusion
table.rename(columns={'class': 'marks'}, inplace = True)
```

```python
table = pd.DataFrame(table)
table
```

# Fixing Table Entities

```python
# Replace values and capitalize column names
table.replace({'Nationality': {'KW': 'Kuwait'},
               'PlaceofBirth': {'KuwaIT': 'Kuwait'},
               'Nationality': {'venzuela': 'Venezuela'},
               'PlaceofBirth': {'venzuela': 'Venezuela'},
               'Stageid': {'lowerlevel': 'PrimarySchool'}
               }, inplace=True)

#Fixing Capitalization
table.columns = [col.capitalize() for col in table.columns]

# Capitalize first letter of each value in selected columns
columns_to_capitalize = ['Nationality', 'Placeofbirth']
for col in columns_to_capitalize:
    table[col] = table[col].str.capitalize()

#Changing KuwaIT to Kuwait
for x in table.index :
    if table.loc[x, 'Placeofbirth'] == 'KuwaIT' :
        table.loc[x, 'Placeofbirth'] = 'Kuwait'
```

```python
#Changing semesters from F-S to 1-2
for x in table.index :
    if table.loc[x, 'Semester'] == 'F' :
        table.loc[x, 'Semester'] = 1
    elif table.loc[x, 'Semester'] == 'S' :
        table.loc[x, 'Semester'] = 2
```

```python
#Changing Absence Days to High or Low
for x in table.index :
    if table.loc[x, 'Studentabsencedays'] == 'Under-7' :
        table.loc[x, 'Studentabsencedays'] = 'Low'
    elif table.loc[x, 'Studentabsencedays'] == 'Above-7' :
        table.loc[x, 'Studentabsencedays'] = 'High'
```

```python
for col in table.columns:
    print(f"Column Name: {col}")
    values = ', '.join(table[col].astype(str).unique())
    print(f"{values}\n")

table
```
