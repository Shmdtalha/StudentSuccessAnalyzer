# Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the order of mark categories
mark_order = ['H', 'M', 'L']

# Iterate over each column
for col in table.columns:
    # Skip 'Marks' column
    if col == 'Marks':
        continue

    # Calculate percentage of each mark category within each group
    counts = table.groupby([col, 'Marks']).size().unstack().fillna(0)
    percentages = counts.div(counts.sum(axis=1), axis=0) * 100

    # Reorder columns to match mark_order
    percentages = percentages[mark_order]

    # Create a new figure for each column
    plt.figure(figsize=(8, 6))

    # Plot barplot for the current column
    percentages.plot(kind='bar', stacked=False)

    # Set plot title
    plt.title(f'Percentage of Marks ({col})')

    # Rotate x-labels for better readability
    plt.xticks(rotation=45)

    # Set y-axis label
    plt.ylabel('Percentage')

    # Show legend
    plt.legend(title='Marks')

    # Show plot
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate percentiles for Raisedhands column
percentiles = table['Raisedhands'].describe(percentiles=[0.25, 0.5, 0.75])

# Define group labels
group_labels = ['0-25%', '25-50%', '50-75%', '75-100%']

# Assign each row to a group based on the percentile value
table['Raisedhands_group'] = pd.cut(table['Raisedhands'], bins=percentiles[['25%', '50%', '75%', 'max']].tolist(), labels=group_labels[:-1], include_lowest=True)

# Iterate over each group
for group in group_labels[:-1]:
    # Filter rows for the current group
    group_df = table[table['Raisedhands_group'] == group]

    # Calculate the percentage of each mark category within the group
    counts = group_df.groupby('Marks').size().fillna(0)
    percentages = counts.div(counts.sum()) * 100

    # Create a new figure for each group
    plt.figure(figsize=(6, 4))

    # Plot pie chart for the current group
    plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=140)

    # Set plot title
    plt.title(f'Distribution of Marks for Raisedhands Group: {group}')

    # Show plot
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Columns to analyze
columns_to_analyze = ['Visitedresources', 'Announcementsview', 'Discussion']

# Iterate over each column
for col in columns_to_analyze:
    print('\n\n' + col)

    # Calculate percentiles for the column
    percentiles = table[col].describe(percentiles=[0.25, 0.5, 0.75])

    # Define group labels
    group_labels = ['0-25%', '25-50%', '50-75%', '75-100%']

    # Assign each row to a group based on the percentile value
    table[f'{col}_group'] = pd.cut(table[col], bins=percentiles[['25%', '50%', '75%', 'max']].tolist(), labels=group_labels[:-1], include_lowest=True)

    # Iterate over each group
    for group in group_labels[:-1]:
        # Filter rows for the current group
        group_df = table[table[f'{col}_group'] == group]

        # Calculate the percentage of each mark category within the group
        counts = group_df.groupby('Marks').size().fillna(0)
        percentages = counts.div(counts.sum()) * 100

        # Create a new figure for each group
        plt.figure(figsize=(6, 4))

        # Plot pie chart for the current group
        plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=140)

        # Set plot title
        plt.title(f'Distribution of Marks for {col} Group: {group}')

        # Show plot
        plt.show()
        print()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the order of 'Marks'
marks_order = ['H', 'M', 'L']

# Select numerical columns for scatter plots
numerical_columns = ['Raisedhands', 'Visitedresources', 'Announcementsview', 'Discussion']

# Create scatter plots for each numerical column against 'Marks', ordered by 'Marks'
for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=table[col], y=table['Marks'], hue=table['Marks'], hue_order=marks_order)
    plt.title(f'Scatter Plot of {col} vs. Marks')
    plt.xlabel(col)
    plt.ylabel('Marks')
    plt.grid(True)
    plt.legend(title='Marks')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Encode 'Marks' column numerically
marks_encoding = {'L': 0, 'M': 1, 'H': 2}
table['Marks_encoded'] = table['Marks'].map(marks_encoding)

# Select numerical columns for correlation analysis
numerical_columns = ['Raisedhands', 'Visitedresources', 'Announcementsview', 'Discussion', 'Marks_encoded']

# Calculate the correlation matrix
correlation_matrix = table[numerical_columns].corr()

# Plot the correlation heatmap including 'Marks'
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap of Numerical Columns including Marks')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Encode 'Marks' column into numerical values
marks_encoding = {'L': 0, 'M': 1, 'H': 2}
table['Marks_encoded'] = table['Marks'].map(marks_encoding)

# Separate data by 'Marks' categories
low_marks_data = table[table['Marks'] == 'L']
medium_marks_data = table[table['Marks'] == 'M']
high_marks_data = table[table['Marks'] == 'H']

# Select numerical columns for correlation analysis
numerical_columns = ['Raisedhands', 'Visitedresources', 'Announcementsview', 'Discussion']

# Calculate correlation for each 'Marks' category
low_corr = low_marks_data[numerical_columns + ['Marks_encoded']].corr()
medium_corr = medium_marks_data[numerical_columns + ['Marks_encoded']].corr()
high_corr = high_marks_data[numerical_columns + ['Marks_encoded']].corr()

# Visualize correlation heatmaps for each 'Marks' category
for marks_category, corr_data in zip(['Low', 'Medium', 'High'], [low_corr, medium_corr, high_corr]):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_data, annot=True, cmap='coolwarm', fmt=".2f", square=True)
    plt.title(f'Correlation Heatmap for {marks_category} Marks')
    plt.show()
```
